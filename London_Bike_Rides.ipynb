```python
import zipfile as zp
import pandas as pd
```

```python
# Unzip and extract the csv file from the folder
zipfile_name = 'London_bike_sharing_dataset.zip'
with zp.ZipFile(zipfile_name, 'r') as file:
    file.extractall()
```

```python
# Read in the csv file as pandas dataframe
bike_rides = pd.read_csv('london_merged.csv')
bike_rides
```

```python
# Renaming the column name
new_cols_dict = {
    'timestamp': 'time',
    'cnt': 'count',
    't1' : 'temp_real_C',
    't2' : 'temp_feels_like_C',
    'hum' : 'humidity_percentage',
    'wind_speed' : 'wind_speed_kph',
    'weather_code' : 'weather',
    'is_holiday' : 'is_holiday',
    'is_weekend' : 'is_weekend',
    'season' : 'season'
}

bike_rides.rename(new_cols_dict, axis=1, inplace=True)
```

```python
# changing the humidity values to percentage (i.e. a value between 0 and 1)
bike_rides.humidity_percentage = bike_rides.humidity_percentage/100
```

```python
# creating the season dictionary so that we can map the integer 0-3 to actual written values
season_dict = {
            '0.0' : 'Spring',
            '1.0' : 'Summer',
            '3.0' : 'Winter',
            '2.0' : 'Autumn'
}

# Creating the waether code values so that we can map the integers to the actual written values
weather_dict = {
                '1.0': 'Clear',
                '2.0': 'Scattered clouds',
                '3.0': 'Broken clouds',
                '4.0': 'Cloudy',
                '7.0': 'Rain',
                '10.0': 'Rain with thunderstom',
                '26.0': 'Snowfall'
}

# change the data tye of season and weather from float to string
bike_rides.season = bike_rides.season.astype('str')
# map the new dict values to season and weather column
bike_rides.season = bike_rides.season.map(season_dict)

# change the data tye of season and weather from float to string
bike_rides.weather = bike_rides.weather.astype('str')
# map the new dict values to season and weather column
bike_rides.weather = bike_rides.weather.map(weather_dict)
                
```
